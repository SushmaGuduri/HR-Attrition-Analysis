### Statistical Test for Job Satisfaction 

```python
# Separate job satisfaction values for employees who left (Attrition == 1) and those who stayed (Attrition == 0)
attrition_yes = df[df['Attrition'] == 1]['JobSatisfaction']  # Attrition == 1 means they left
attrition_no = df[df['Attrition'] == 0]['JobSatisfaction']   # Attrition == 0 means they stayed

# Perform an independent t-test to compare the means of job satisfaction for both groups
t_stat, p_value = stats.ttest_ind(attrition_yes, attrition_no)

# Print the t-statistic and p-value
print('T-test for Job Satisfaction between Attrition groups:')
print('T-statistic:', t_stat)
print('P-value:', p_value)
```

```python
# Interpretation:
# If p-value < 0.05, we reject the null hypothesis, meaning there is a statistically significant difference in job satisfaction between employees who left and those who stayed.
if p_value < 0.05:
    print('Conclusion: There is a significant difference in Job Satisfaction between those who left and those who stayed.')
else:
    print('Conclusion: There is no significant difference in Job Satisfaction between the two groups.')
```

## Attrition by Department

```python
# Chi-square test for Attrition by Department
contingency_table = pd.crosstab(df['Department'], df['Attrition'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print('Chi-square test for Department:', chi2, p)
```

```python
# Interpretation:
# If p-value < 0.05, we reject the null hypothesis, meaning that attrition is dependent on department.
if p < 0.05:
    print('Conclusion: Attrition is dependent on Department.')
else:
    print('Conclusion: Attrition is independent of Department.')
```

```python
# Calculate attrition rate by department
attrition_rate = df.groupby('Department')['Attrition'].mean() * 100
print("Attrition Rate by Department:")
print(attrition_rate)
```

```python
import matplotlib.pyplot as plt

# Bar chart to visualize attrition by department
attrition_rate.plot(kind='bar', color='skyblue')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

## Job Satisfaction and Log Monthly Income: High vs Low Attrition Departments

```python
# Compare Job Satisfaction in departments with high attrition
high_attrition_departments = attrition_rate[attrition_rate > attrition_rate.mean()].index
low_attrition_departments = attrition_rate[attrition_rate <= attrition_rate.mean()].index

# Analyze JobSatisfaction and Log_MonthlyIncome for high vs low attrition departments
job_satisfaction_high = df[df['Department'].isin(high_attrition_departments)]['JobSatisfaction'].mean()
job_satisfaction_low = df[df['Department'].isin(low_attrition_departments)]['JobSatisfaction'].mean()

# Use the log-transformed MonthlyIncome
monthly_income_high = df[df['Department'].isin(high_attrition_departments)]['Log_MonthlyIncome'].mean()
monthly_income_low = df[df['Department'].isin(low_attrition_departments)]['Log_MonthlyIncome'].mean()

print(f"Job Satisfaction in High Attrition Departments: {job_satisfaction_high}")
print(f"Job Satisfaction in Low Attrition Departments: {job_satisfaction_low}")
print(f"Log Monthly Income in High Attrition Departments: {monthly_income_high}")
print(f"Log Monthly Income in Low Attrition Departments: {monthly_income_low}")
```

```python
# Data
categories = ['High Attrition', 'Low Attrition']
job_satisfaction_values = [job_satisfaction_high, job_satisfaction_low]
log_income_values = [monthly_income_high, monthly_income_low]

# Create a bar plot for Job Satisfaction
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(categories, job_satisfaction_values, color=['red', 'green'])
plt.title('Job Satisfaction in High vs Low Attrition Departments')
plt.ylabel('Average Job Satisfaction')
plt.ylim(1.8, 2.0)

# Create a bar plot for Log Monthly Income
plt.subplot(1, 2, 2)
plt.bar(categories, log_income_values, color=['blue', 'orange'])
plt.title('Log Monthly Income in High vs Low Attrition Departments')
plt.ylabel('Log Monthly Income')
plt.ylim(8.4, 8.7)

plt.tight_layout()
plt.show()
```

## Cluster Analysis and Insights

```python
from sklearn.cluster import KMeans

# Selecting relevant numeric features for clustering
features = df[['Age', 'Log_MonthlyIncome', 'JobSatisfaction']]

# Apply KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(features)

# Visualizing the clusters using scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Log_MonthlyIncome', hue='Cluster', palette='Set1', data=df)
plt.title('Clusters of Employees based on Age and Log Monthly Income')
plt.show()

# Cluster insights
cluster_means = df.groupby('Cluster')[['Age', 'Log_MonthlyIncome', 'JobSatisfaction']].mean()
print("Cluster Centers:")
print(cluster_means)
```
